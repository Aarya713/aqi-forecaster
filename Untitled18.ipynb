```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Conv1D, MaxPooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import keras_tuner as kt
import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load AQI data
aqi_df = pd.read_csv('final_dataset.csv')
aqi_df['Date'] = pd.to_datetime(aqi_df[['Year', 'Month', 'Date']].rename(columns={'Date': 'Day'}))
aqi_df.set_index('Date', inplace=True)
aqi_df.sort_index(inplace=True)

# Load weather data
weather_df = pd.read_csv('india_2000_2024_daily_weather.csv', parse_dates=['date'])
weather_df.set_index('date', inplace=True)
weather_df.sort_index(inplace=True)

# Filter for Delhi
weather_df = weather_df[weather_df['city'].str.lower() == 'delhi']
weather_df.drop(columns=['city'], inplace=True)

# Merge
merged_df = aqi_df.join(weather_df, how='inner')
print("Merged shape:", merged_df.shape)
```

```python
# Merge
merged_df = aqi_df.join(weather_df, how='inner')
print("Merged shape:", merged_df.shape)

if 'wind_direction_10m_dominant' in merged_df.columns:
    merged_df['wind_dir_sin'] = np.sin(np.radians(merged_df['wind_direction_10m_dominant']))
    merged_df['wind_dir_cos'] = np.cos(np.radians(merged_df['wind_direction_10m_dominant']))
```

```python
# Ensure you have merged_df in memory (from earlier cells)
# Extract AQI series for the last 30 days relative to today
# We'll use the most recent dates available in merged_df (which goes up to 2024)

def get_historical_aqi_for_dates(dates):
    """
    Given a list of dates (pandas DatetimeIndex), return a Series of AQI values
    from merged_df for those dates. If a date is not found, forward fill from the previous day.
    """
    # merged_df has date index, we'll reindex to the requested dates
    aqi_series = merged_df['AQI']
    # Reindex to the requested dates, forward fill missing
    return aqi_series.reindex(dates, method='ffill')
```

```python
if 'wind_direction_10m_dominant' in merged_df.columns:
    merged_df['wind_dir_sin'] = np.sin(np.radians(merged_df['wind_direction_10m_dominant']))
    merged_df['wind_dir_cos'] = np.cos(np.radians(merged_df['wind_direction_10m_dominant']))
```

```python
rain_threshold = 0.1
merged_df['rain_event'] = (merged_df['precipitation_sum'] > rain_threshold).astype(int)
merged_df['days_since_rain'] = merged_df.groupby(merged_df['rain_event'].cumsum()).cumcount()
```

```python
merged_df['is_weekend'] = merged_df['Days'].isin([6,7]).astype(int)
```

```python
month_dummies = pd.get_dummies(merged_df['Month'], prefix='month')
merged_df = pd.concat([merged_df, month_dummies], axis=1)
```

```python
for lag in [1, 2, 3, 7]:
    merged_df[f'AQI_lag_{lag}'] = merged_df['AQI'].shift(lag)
```

```python
for window in [3, 7, 30]:
    merged_df[f'AQI_roll_mean_{window}'] = merged_df['AQI'].rolling(window).mean()
    merged_df[f'AQI_roll_std_{window}'] = merged_df['AQI'].rolling(window).std()
```

```python
alpha = 0.3
merged_df['PM25_ewma'] = merged_df['PM2.5'].ewm(alpha=alpha, adjust=False).mean()
```

```python
merged_df['rain_wind_interaction'] = merged_df['precipitation_sum'] * merged_df['wind_speed_10m_max']
```

```python
print("Shape before dropping NaNs:", merged_df.shape)
merged_df.dropna(inplace=True)
print("Shape after dropping NaNs:", merged_df.shape)
```

```python
target = 'AQI'
feature_cols = [col for col in merged_df.columns if col != target]

train_size = int(len(merged_df) * 0.8)
train_df = merged_df.iloc[:train_size]
test_df = merged_df.iloc[train_size:]

print(f"Train rows: {len(train_df)}, Test rows: {len(test_df)}")
```

```python
# Use entire training set as flat features
X_train_flat = train_df[feature_cols].values
y_train_flat = train_df[target].values

rf_selector = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_selector.fit(X_train_flat, y_train_flat)

importances = rf_selector.feature_importances_
indices = np.argsort(importances)[::-1]

# Select top K features
K = 100
selected_features = [feature_cols[i] for i in indices[:K]]
print(f"Selected top {K} features (first 10):", selected_features[:10])

# Restrict dataframes
train_df_sel = train_df[selected_features + [target]]
test_df_sel = test_df[selected_features + [target]]
```

```python
def prepare_sequences(train_df, test_df, feature_cols, target, window_size):
    X_train = train_df[feature_cols].values
    y_train = train_df[target].values.reshape(-1,1)
    X_test = test_df[feature_cols].values
    y_test = test_df[target].values.reshape(-1,1)

    scaler_X = MinMaxScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)

    scaler_y = MinMaxScaler()
    y_train_scaled = scaler_y.fit_transform(y_train)
    y_test_scaled = scaler_y.transform(y_test)

    def create_sequences(X, y, ws):
        X_seq, y_seq = [], []
        for i in range(ws, len(X)):
            X_seq.append(X[i-ws:i, :])
            y_seq.append(y[i, 0])
        return np.array(X_seq), np.array(y_seq)

    X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train_scaled, window_size)
    X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test_scaled, window_size)

    return X_train_seq, y_train_seq, X_test_seq, y_test_seq, scaler_y
```

```python
def prepare_sequences(train_df, test_df, feature_cols, target, window_size):
    X_train = train_df[feature_cols].values
    y_train = train_df[target].values.reshape(-1,1)
    X_test = test_df[feature_cols].values
    y_test = test_df[target].values.reshape(-1,1)

    scaler_X = MinMaxScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)

    scaler_y = MinMaxScaler()
    y_train_scaled = scaler_y.fit_transform(y_train)
    y_test_scaled = scaler_y.transform(y_test)

    def create_sequences(X, y, ws):
        X_seq, y_seq = [], []
        for i in range(ws, len(X)):
            X_seq.append(X[i-ws:i, :])
            y_seq.append(y[i, 0])
        return np.array(X_seq), np.array(y_seq)

    X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train_scaled, window_size)
    X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test_scaled, window_size)

    return X_train_seq, y_train_seq, X_test_seq, y_test_seq, scaler_y
```

```python
window_size = 30
X_tr, y_tr, X_te, y_te, scaler_y = prepare_sequences(
    train_df_sel, test_df_sel, selected_features, target, window_size
)

def build_model(hp):
    model = Sequential()
    # First LSTM
    model.add(LSTM(
        units=hp.Int('units1', 32, 128, step=32),
        return_sequences=True,
        input_shape=(X_tr.shape[1], X_tr.shape[2])
    ))
    model.add(Dropout(hp.Float('dropout1', 0.2, 0.4, step=0.1)))

    # Second LSTM
    model.add(LSTM(
        units=hp.Int('units2', 32, 128, step=32),
        return_sequences=True
    ))
    model.add(Dropout(hp.Float('dropout2', 0.2, 0.4, step=0.1)))

    # Optional third LSTM
    if hp.Boolean('third_layer'):
        model.add(LSTM(
            units=hp.Int('units3', 32, 128, step=32),
            return_sequences=False
        ))
        model.add(Dropout(hp.Float('dropout3', 0.2, 0.4, step=0.1)))
    else:
        model.add(LSTM(
            units=hp.Int('units_last', 32, 128, step=32)
        ))
        model.add(Dropout(hp.Float('dropout_last', 0.2, 0.4, step=0.1)))

    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1))

    model.compile(
        optimizer=Adam(learning_rate=hp.Choice('lr', [0.001, 0.0005])),
        loss='mse'
    )
    return model

tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=20,
    executions_per_trial=2,
    directory='tuner_results',
    project_name='aqi_lstm'
)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

tuner.search(X_tr, y_tr,
             epochs=50,
             batch_size=32,
             validation_data=(X_te, y_te),
             callbacks=[early_stop],
             verbose=1)

# Get best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best hyperparameters:")
print(best_hps.values)

# Build the best model
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
window_size = 30
X_tr, y_tr, X_te, y_te, scaler_y = prepare_sequences(
    train_df_sel, test_df_sel, selected_features, target, window_size
)
```

```python
y_pred_scaled = best_model.predict(X_te)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_actual = scaler_y.inverse_transform(y_te.reshape(-1,1))

rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
mae = mean_absolute_error(y_actual, y_pred)
r2 = r2_score(y_actual, y_pred)
print(f"Tuned LSTM - RMSE: {rmse:.2f}, MAE: {mae:.2f}, R²: {r2:.4f}")
```

```python
# Recreate scaler_X from the original training data
scaler_X = MinMaxScaler()
scaler_X.fit(train_df_sel[selected_features].values)

# Now save everything
best_model.save('best_aqi_model.h5')          # you can also use .keras to avoid the warning
joblib.dump(scaler_X, 'scaler_X.pkl')
joblib.dump(scaler_y, 'scaler_y.pkl')
joblib.dump(selected_features, 'selected_features.pkl')
print("Model and scalers saved.")
```
